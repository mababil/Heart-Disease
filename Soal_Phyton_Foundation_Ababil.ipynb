# 2. Hitunglah BMI setiap siswa dengan rumus BMI=Berat badan (kg)(Tinggi Badan (m))2

```python
# 2
def bmi_calc(berat, tinggi):
  return berat / (tinggi/100)**2
print(df)
```

# 4. Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.



```python
# 3 & 4
list_bmi = []
for i in range(len(df)):
  berat = df["Berat Badan (kg)"][i]
  tinggi = df["Tinggi Badan (cm)"][i]
  bmi = bmi_calc(berat, tinggi)
  list_bmi.append(bmi)
```

```python
df["BMI"] = list_bmi
df
```

```python
  # F
def get_bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi <= 24.9:
        return 'Normal'
    elif 25.0 <= bmi <= 29.9:
        return 'Overweight'
    elif 30.0 <= bmi <= 34.9:
        return 'Obesity'
    else:
        return 'Unknown'

# Terapkan fungsi untuk mendapatkan kategori BMI untuk setiap siswa
df['BMI Category'] = df['BMI'].apply(get_bmi_category)

# Hitung jumlah siswa berdasarkan kategori BMI
status_count = df['BMI Category'].value_counts()

# Tampilkan jumlah siswa berdasarkan kategori BMI dengan keterangan berat badan
print("\nJumlah siswa berdasarkan kategori BMI:")
for status, count in status_count.items():
    print(f"{status}: {count} siswa")
```

## SOAL 2

 Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:
Input
>> Money Capital : 100000
>> Investment Tenure (years) : 2
Output
>> Year 1, your total investment : 125000.0
>> Year 2, your total investment : 156250.0

```python
def main():
    # Menerima input modal investasi dari pengguna
    while True:
        money_capital = input("Money Capital: ")
        if money_capital.isdigit() and 0 <= int(money_capital) <= 1000000000:
            money_capital = int(money_capital)
            break
        else:
            print("Modal investasi harus berupa bilangan bulat antara 0-1.000.000.000.")

    # Menerima input jangka waktu investasi dari pengguna
    while True:
        investment_tenure = input("Investment Tenure (years): ")
        if investment_tenure.isdigit() and 1 <= int(investment_tenure) <= 25:
            investment_tenure = int(investment_tenure)
            break
        else:
            print("Jangka waktu investasi harus berupa bilangan bulat antara 1-25 dalam tahun.")

    # Hitung dan tampilkan akumulasi investasi setiap tahun
    total_investment = money_capital
    for year in range(1, investment_tenure + 1):
        total_investment *= 1.25
        print(f"Year {year}, your total investment: {total_investment:.1f}")

if __name__ == "__main__":
    main()
```
